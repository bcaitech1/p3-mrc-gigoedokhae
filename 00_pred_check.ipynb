```python
import json
from datasets import load_from_disk
import pandas as pd

# predictions
def get_preds(path):
    with open(path) as f:
        return json.load(f) # dict
paths, preds = [], []
paths.append("/opt/ml/outputs/preds/backup/no1_xlm_roberta/predictions.json")
paths.append("/opt/ml/outputs/preds/no16_no6_xlm_roberta_fine_tuning_continue_checkpoint-8820/predictions.json")
for path in paths:
    preds.append(get_preds(path))

# test dataset(to DataFrame)
path = "/opt/ml/input/data/test_dataset"
dataset = load_from_disk(path)["validation"]
df = pd.DataFrame(dataset)
```

```python
from IPython.display import display
def str_formatter(string, pp, cycle):
    pp.text(string)
plain = get_ipython().display_formatter.formatters['text/plain']
plain.for_type(str, str_formatter)


displayer = display("", display_id=True)

diff_cnt = 0
for q_id, *answers in zip(preds[0], *(pred.values() for pred in preds)):
    if len(set(answers)) != 1:
        diff_cnt += 1
        print(df["question"][df["id"] == q_id].values[0])
        for answer in answers:
            print(f"- {answer}")
displayer.update(f"*다른 답의 개수: {diff_cnt}\n\n")
```
